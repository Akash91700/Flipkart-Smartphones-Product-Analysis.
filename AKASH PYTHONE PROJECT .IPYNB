# Smartphone Price Segment

- Budget (≤ ₹12,000) → Entry-level phones, basic specs, good for first-time buyers.
- Affordable Mid-Range (₹12,001 – ₹20,000) → Balanced performance, 4–6 GB RAM, decent cameras.
- Performance Mid-Range (₹20,001 – ₹35,000) → Gaming + photography focus, AMOLED displays, fast charging.
- Upper Mid-Range (₹35,001 – ₹55,000) → Premium features, 108MP cameras, 5G, high refresh rate screens.
- Flagship (₹55,001 – ₹80,000) → Top-tier processors, pro-grade cameras, premium build.
- Ultra-Flagship (₹80,001+) → Luxury devices, foldables, iPhone Pro Max, Galaxy Z series.


```python

# Function to classify phones based on price segment
def classify_segment(price):
    if price <= 12000:
        return "Budget"                  # Low-cost phones
    elif 12001 <= price <= 20000:
        return "Affordable Mid-Range"    # Entry mid-range
    elif 20001 <= price <= 35000:
        return "Performance Mid-Range"   # Balanced performance
    elif 35001 <= price <= 55000:
        return "Upper Mid-Range"         # Premium mid-range
    elif 55001 <= price <= 80000:
        return "Flagship"               # High-end flagship
    else:
        return "Ultra-Flagship"         # Top-tier ultra premium
```

```python
df["price_segment"] = df["original_price"].apply  (classify_segment)    # Apply the classify_segment function to 'original_price' column

```

- Insight → Memory (RAM + Storage) grows together, but battery capacity is segment-driven (budget vs flagship) rather than tied to memory specs.
